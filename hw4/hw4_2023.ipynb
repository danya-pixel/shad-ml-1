Пусть $X_1,...,X_n$ &mdash; выборка из экспоненциального распределения с плотностью $p_\theta(x) = \theta e^{-\theta x},$ где $x>0, \theta>0$. Такое распределение описывает время до наступления некоторого события. Например, таким событием может быть поступление очередного заказа таксисту или курьеру службы доставки. В таком случае $X_i$ &mdash; время ожидания $i$-го заказа после завершения предыдущего. Параметр $\theta$ имеет смысл интенсивности заказов в единицу времени. Следовательно, чем больше $\theta$, тем меньше среднее значение $X_i$. Точнее $\mathsf{E}_\theta X_i = 1/\theta$.


При выполнении задач используйте то, что $\mathsf{E}_\theta X_i^k = k!/\theta^k$.

Запишем ЦПТ:

$$
\sqrt{ n }(\overline{X} - EX_{i}) \xrightarrow{d} N(0, \sigma^2)
$$
Знаем, что $EX_{i} = \frac{1}{\theta}$, тогда

$$
\sqrt{ n }\left( \overline{X} - \frac{1}{\theta} \right) \xrightarrow{d} N(0, \sigma^2)
$$
Определение А.Н.О. – $\sqrt{ n }(\hat{\theta} - \theta) \to N(0, \sigma^2(\theta))$
Тогда, $\overline{X}$ является А.Н.О. для $\frac{1}{\theta}$. 

Применим дельта-метод: пусть $h(x) = \frac{1}{x}$, тогда $[h'(x)]^2=\frac{1}{x^4}$. Поставляем $\frac{1}{\theta}$ => $\left[ h'\left( \frac{1}{\theta} \right) \right]^2=\theta^4$.

Найдем асимптотическую дисперсию. Из формулы дельта-метода:
$$
\dots \xrightarrow{d}N(0, \sigma^2[h'(\theta)]^2)
$$
В нашем случае $\sigma^2=DX_{1}=\frac{1}{\theta^2}$. Получаем асимптотическую дисперсию:
$$
\frac{\theta^4}{\theta^2} =\theta^2.
$$

Получили А.Н.О. $\frac{1}{\overline{X}}$ с асимптотической дисперсией $\theta^2$.

**1.** Сначала вспомним, как следует интерпретировать доверительные интервалы и статистические критерии. Зафиксируем значение $\theta$, соответствующее гипотезе $\mathsf{H}_0$.

```python
theta = 1
```

Давайте промоделируем множество экспериментов с одинаковыми условиями. Сгенерируйте порядка 20 выборок размера 10, для каждой из них
* посчитайте численно границы всех трех доверительных интервалов
* проверьте гипотезу статистическими критериями.

Результат можно удобно оформить в виде таблицы, по строкам которой будут эксперименты, а по столбцам &mdash; границы доверительных интервалов и результаты применения критериев.

```python
import numpy as np
import scipy.stats as sps
import pandas as pd

n = 100000
num_experiments = 20
statistics = []
alpha = 0.95

exp_samples = [sps.expon(scale=1 / theta).rvs(size = n) for _ in range(num_experiments)]
```

Для оценки $\frac{1}{\overline{X}}$:

```python
for samples in exp_samples:

    quantile = sps.norm.ppf((1 + alpha) / 2)
    samples_mean = np.mean(samples)
    estimate = 1/samples_mean

    left_border = estimate - quantile*(estimate)/(np.sqrt(n))
    right_border = estimate + quantile*(estimate)/(np.sqrt(n))
    statistics.append([left_border, right_border])

stats = pd.DataFrame(statistics, columns=['left_border_1', 'right_border_1'])
```

Для оценки $\sqrt{ \frac{2}{\overline{X^2}} }$ 

```python
statistics = []

for samples in exp_samples:

    quantile = sps.norm.ppf((1 + alpha) / 2)

    samples_mean = np.mean(np.power(samples, 2))
    estimate = np.sqrt(2/samples_mean)
    sigma = np.sqrt(5/(2*samples_mean))

    left_border = estimate - quantile*(sigma)/(np.sqrt(n))
    right_border = estimate + quantile*(sigma)/(np.sqrt(n))
    statistics.append([left_border, right_border])

stats_2 = pd.DataFrame(statistics, columns=['left_border_2', 'right_border_2'])
stats = pd.concat([stats, stats_2], axis=1)
```

Для оценки $\overline{X}$:

```python
statistics = []

for samples in exp_samples:
    quantile = sps.norm.ppf((1 + alpha) / 2)

    gamma_distribution = sps.gamma(a=n, scale=1 / n)
    samples_mean = np.mean(samples)
    estimate = samples_mean
    
    left_border = gamma_distribution.ppf((1-alpha)/2)/estimate
    right_border = gamma_distribution.ppf((1+alpha)/2)/estimate

    statistics.append([left_border, right_border])

stats_3 = pd.DataFrame(statistics, columns=['left_border_3', 'right_border_3'])
stats = pd.concat([stats, stats_3], axis=1)
```

Интервалы с дотаточной вероятностью **покрывают** реализации выборки. Однако, при таких малых N сложно сказать о том, что какой-то из интервалов лучше/хуже. 

**2.** Для доверительных интервалов повторите эксперимент, показанный на семинаре, 
* построив визуализации доверительны интервалов при увеличении размера выборки,

Кроме этого, оцените и визуализируйте также среднюю длину каждого интервала в зависимости от размера выборки.

```python
import matplotlib.pyplot as plt

def draw_confidence_interval(
    left,
    right,
    estimation=None,
    sample=None,
    ylim=(None, None),
    color_estimation='#FF3300',
    color_interval='#00CC66',
    color_sample='#0066FF',
    label_estimation='',
    label_interval='',
    legend=None,
):
    '''
    Рисует доверительный интервал и оценку в зависимости от размера выборки.

    :param left: левые границы интервалов (в зависимости от n)
    :param right: правые границы интервалов (в зависимости от n)
    :param estimation: оценки (в зависимости от n)
    :param sample: выборка
    :param ylim: ограничение вертикальной оси
    :param color_estimation: цвет оценки
    :param color_interval: цвет интервала
    :param color_sample: цвет выборки
    :param label_estimation: подпись для оценки
    :param label_interval: подпись для интервала
    '''

    time = np.arange(len(left)) + 1

    assert len(time) == len(right)

    if not sample is None:
        plt.scatter(time, sample, alpha=0.3, color=color_sample, s=40, label='Элемент выборки')
    if not estimation is None:
        plt.plot(time, estimation, color=color_estimation, linewidth=2.5, label=label_estimation)

    plt.fill_between(time, left, right, alpha=0.25, color=color_interval, label=label_interval)

    plt.legend()
    plt.ylim(ylim)
    plt.xlabel('Размер выборки')
    if legend is not None:
        plt.title(legend)
    else:
        plt.title('Зависимость доверительного интервала от размера выборки')
```

Что можно теперь сказать про эти доверительные интервалы?

Точный интервал плохо работает при малых размерах выборки, т.к. сильно опирается на среднее. Интервал по оценке 1 имеет меньшую дисперсию чем интервал по оценке 2 (что также следует из формул и значения асимптотической дисперсии). При больших N все интервалы сходятся.

**3.** Для всех исследуемых критериев оцените вероятность ошибки 1 рода и постройте для них доверительный интервал. Пример оценки стоит посмотреть в семинарском ноутбуке, но в данном задании требуется построить такую оценку в засимимости от размера выборки.

```python
def conf_int_1(sample_len, sample_sum, alpha=0.95):
    quantile = sps.norm.ppf((1+alpha)/2)

    samples_mean = sample_sum/sample_len
    estimate = 1/samples_mean

    left_border = estimate - quantile*(estimate)/(np.sqrt(sample_len))
    right_border = estimate + quantile*(estimate)/(np.sqrt(sample_len))
    return (left_border, right_border)

def crit_1(X, alpha=0.05):
    """
    Функция реализует описанный выше критерий.
    X: выборка
    alpha: желаемая ошибка первого рода
    """
    C = conf_int_1(len(X), np.sum(X), alpha=1-alpha)
    return not(C[0] <= 1 <= C[1])

def conf_int_error(sample_len, sample_sum, alpha=0.95):
    """
    функция возвращает (L, R) - доверительный интервал уровня доверия 
    param: alpha 
        по выборке 
    param: X
    """
    
    p_hat = sample_sum / sample_len
    sigma = np.sqrt((p_hat * (1 - p_hat)) / sample_len)
    q = sps.norm.ppf((1 + alpha) / 2)
    C = (p_hat - q * sigma, p_hat + q * sigma)
    return C


def calculate_1error(crit, n_iter=10000, n=1000): 
    """
    функция оценивает вероятность ошибки первого рода
    crit: критерий
    n_iter: число итераций оценивания
    n: размер выборки
    возвращает точечную оценку и доверительный интервал
    """
    n_reject = 0  # число отвержений
    
    for _ in range(n_iter):
        X = sps.expon(scale=1).rvs(size = n) 
        n_reject += crit(X)
    
    C = conf_int_error(n_iter, n_reject)
    return n_reject / n_iter, C
```

```python
calculate_1error(crit_1)
```

```python
def draw_error_confidence_interval(crit, num, sizes=np.arange(1, 1000, 50)):
    means = []
    left = []
    right = []
    for i in sizes:
        m, (l, r) = calculate_1error(crit, n=i)
        means.append(m)
        left.append(l)
        right.append(r)

    draw_confidence_interval(
        left,
        right,
        means,
        ylim=(None, None),
        color_estimation='#E35BA1',
        color_interval='#FF3300',
        label_interval='Интервал ошибки 1го рода',
        legend=f'Ошибка 1го рода для оценки {num}',
    )

draw_error_confidence_interval(crit_1, 1)
```

```python
def conf_int_2(sample_len, sample_sum, alpha=0.95):
    quantile = sps.norm.ppf((1 + alpha) / 2)

    samples_mean = sample_sum/sample_len

    estimate = np.sqrt(2/samples_mean)
    sigma = np.sqrt(5/(2*samples_mean))

    left_border = estimate - quantile*(sigma)/(np.sqrt(sample_len))
    right_border = estimate + quantile*(sigma)/(np.sqrt(sample_len))
    return (left_border, right_border)

def crit_2(X, n, alpha=0.05):
    """
    Функция реализует описанный выше критерий.
    X: выборка
    alpha: желаемая ошибка первого рода
    """
    C = conf_int_2(len(X), np.sum(np.power(X, 2)[:n]), 1-alpha)
    return not(C[0] <= 1 <= C[1])
```

```python
def conf_int_3(sample_len, sample_sum, alpha=0.95):
    gamma_distribution = sps.gamma(a=sample_len, scale=1 / sample_len)
    samples_mean = sample_sum/sample_len
    estimate = samples_mean
    
    left_border = gamma_distribution.ppf((1-alpha)/2)/estimate
    right_border = gamma_distribution.ppf((1+alpha)/2)/estimate
    return (left_border, right_border)

def crit_3(X, alpha=0.05):
    """
    Функция реализует описанный выше критерий.
    X: выборка
    alpha: желаемая ошибка первого рода
    """
    C = conf_int_3(len(X), np.sum(X), 1-alpha)
    return not(C[0] <= 1 <= C[1])
```

Какие критерии не контролируют вероятность ошибки 1 рода на уровне 5\%?

Сложно сказать, у каждого критерия доверительный интервал ошибки 1го рода слегка выходит за 5%. 1й и 2й критерий в среднем показывают результаты лучше, чем 3й. 

Оцените мощность критериев в зависимости от значения $\theta$ и размера выборки. Пример такой оценки есть в семинарском ноутбуке.

```python
from tqdm import tqdm
import seaborn as sns

def calculate_2error(crit, n_iter=10000):
        """
        функция оценивает вероятность ошибки 2го рода для p=param:p и возвращает 
            число от 0 до 1 - долю не отвергнутых гипотез
        crit: критерий
        n_iter: число итераций оценивания
        n: размер выборки
        """
        n_pos = 0 # число НЕотвержений

        for _ in range(n_iter):
            X = sps.expon(scale=1).rvs(size=1000) 
            n_pos += 1 - crit(X)
        return n_pos / n_iter

def calculate_power(crit=crit_1, n_iter=10000):
    """
    функция возвращает мощность критерия для рассматриваемого эксперимента
    param crit: критерий
    param n: размер выборки
    param n_iter: число итераций
    """
    return 1 - calculate_2error(crit, n_iter=n_iter)


def draw_power(crits, show=True):
    """
    Функция рисует график вероятности ошибки второго рода критерия
        param:crit в зависимости от p
    """
    N = [10, 50, 100, 250, 500] 
    for name, crit in crits:
        error_line = [] 
        
        for n in tqdm(N):
            error_line.append(calculate_power(crit, n_iter=10000))
        
        sns.lineplot(x=N, y=error_line, label=name, lw=3)
    plt.xlabel("Размер выборки")
    plt.ylabel("Мощность")
    plt.title("Мощность в завивимости от N")
    if show:
        plt.show()
```

```python
c1 = calculate_power(crit_1)
c2 = calculate_power(crit_2)
c3 = calculate_power(crit_3)

print(f"Мощность критерия 1: {c1}")
print(f"Мощность критерия 2: {c2}")
print(f"Мощность критерия 3: {c3}")
```

Сперва, для работы с формулой предсказательного интервала занесем свободный коэфициент в X.

```python
df['intercept'] = np.ones((len(df), 1))

mode = f"price ~ {' + '.join(df.columns[1:])} -1"

linear = smf.ols(mode, df)
l_results = linear.fit()
print(l_results.summary())
```

```python
ideal = np.array([66, 50, 10, 10, 5, 1, 1, 1, 1, 1])
n, d = len(df), len(df.columns[1:])

X = df.drop('price', axis=1).to_numpy()
y = df['price'].to_numpy()

def predictional_interval(x0, X, y, results):
    cov = pd.DataFrame(results.cov_params(), 
    index=l_results.params.index, 
    columns=l_results.params.index
    ).to_numpy()

    theta_hat = np.array(results.params)
    predictions = x0.T.dot(theta_hat) 
    sigma = np.sqrt(1/(n-d)*np.sum(np.square(y-X.dot(theta_hat))))
    quant_term = sps.t.ppf((1+alpha)/2, n-d)*sigma*np.sqrt(1+x0.T.dot(cov).dot(x0)/np.square(sigma))
    right = predictions + quant_term 
    left = predictions - quant_term
    return predictions, (left, right) 

def response_interval(x0, X, results):
    cov = pd.DataFrame(results.cov_params(), 
    index=l_results.params.index, 
    columns=l_results.params.index
    ).to_numpy()

    theta_hat = np.array(results.params)
    predictions = x0.T.dot(theta_hat) 
    quant_term = sps.t.ppf((1+alpha)/2, n-d)*np.sqrt(x0.T.dot(cov).dot(x0))
    right = predictions + quant_term 
    left = predictions - quant_term
    return predictions, (left, right)
    
robust_results = l_results.get_robustcov_results(cov_type="HC3")
predictional_interval(ideal, X, y, robust_results)
```

Зафиксируйте все параметры, кроме площади квартиры и постройте доверительную ленту для цены квартиры. Иначе говоря, для каждого значения площади квартиры необходимо построить доверительный интервал для среднего значения цены квартиры. Аналогичным образом постройте предсказательную ленту.

```python
surfs = [30, 40, 50, 60, 70, 80, 90, 100, 120, 150]
ll, rr = [], []
pl, pr = [], []
preds = []
for surf in surfs:
    ideal = np.array([surf, 50, 10, 10, 5, 1, 1, 1, 1, 1])
    pred, (response_l, response_r) = response_interval(ideal, X, robust_results)
    _, (pred_l, pred_r) = predictional_interval(ideal, X, y, robust_results)
    print(f"P={surf}, predict={pred:0.2f}, ДИ ({response_l:0.2f}, {response_r:0.2f}), ПИ  ({pred_l:0.2f}, {pred_r:0.2f})")
    preds.append(pred)
    ll.append(response_l)
    rr.append(response_r)

    pl.append(pred_l)
    pr.append(pred_r)
```
